```python
class Customer:
    def __init__(self, name, salary, loan):
        self.name = name
        self.salary = salary
        self.loan = loan

class Employee:
    def __init__(self, name, salary, position):
        self.name = name
        self.salary = salary
        self.position = position

class InformationSystem:
    def __init__(self):
        self.customers = []
        self.employees = []

    def add_customer(self):
        name = input("Enter customer name: ")
        salary = float(input("Enter customer salary: "))
        loan = float(input("Enter loan amount: "))

        customer = Customer(name, salary, loan)
        self.customers.append(customer)
        print("Customer information added successfully.")

    def add_employee(self):
        name = input("Enter employee name: ")
        salary = float(input("Enter employee salary: "))
        position = input("Enter employee position: ")

        employee = Employee(name, salary, position)
        self.employees.append(employee)
        print("Employee information added successfully.")

    def display_customer_info(self):
        print("Customer Information:")
        for customer in self.customers:
            print(f"Name: {customer.name}")
            print(f"Salary: {customer.salary}")
            print(f"Loan: {customer.loan}")
            print("")

    def display_employee_info(self):
        print("Employee Information:")
        for employee in self.employees:
            print(f"Name: {employee.name}")
            print(f"Salary: {employee.salary}")
            print(f"Position: {employee.position}")
            print("")

    def run(self):
        while True:
            print("Options:")
            print("1. Add Customer")
            print("2. Add Employee")
            print("3. Display Customer Information")
            print("4. Display Employee Information")
            print("5. Quit")

            choice = input("Enter your choice (1-5): ")

            if choice == '1':
                self.add_customer()
            elif choice == '2':
                self.add_employee()
            elif choice == '3':
                self.display_customer_info()
            elif choice == '4':
                self.display_employee_info()
            elif choice == '5':
                break
            else:
                print("Invalid choice. Please try again.")

        print("Program exited.")

# Create an instance of InformationSystem and run the program
info_system = InformationSystem()
info_system.run()
```
